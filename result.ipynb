```python
import matplotlib.pyplot as plt
import sys
```

```python
## one-hot encoding 결과 불러오기
```

```python
negative_f = open("result/Negative(OneHotEncoding).txt", "r")
gaussian_noise_f = open("result/GaussianNoise(OneHotEncoding).txt", "r")
gaussian_blur_f = open("result/GaussianBlur(OneHotEncoding).txt", "r")
salt_and_pepper_f = open("result/SaltAndPepper(OneHotEncoding).txt", "r")
fgsm_f = open("result/FGSM(OneHotEncoding).txt", "r")

negative_one_hot = negative_f.read().split("\n")
negative_acc_one_hot = eval(negative_one_hot[0])

gaussian_noise_one_hot = gaussian_noise_f.read().split("\n")
gaussian_noise_var_one_hot = eval(gaussian_noise_one_hot[0])
gaussian_noise_acc_one_hot = eval(gaussian_noise_one_hot[1])

gaussian_blur_one_hot = gaussian_blur_f.read().split("\n")
gaussian_blur_sigma_one_hot = eval(gaussian_blur_one_hot[0])
gaussian_blur_acc_one_hot = eval(gaussian_blur_one_hot[1])

salt_and_pepper_one_hot = salt_and_pepper_f.read().split("\n")
salt_and_pepper_amount_one_hot = eval(salt_and_pepper_one_hot[0])
salt_and_pepper_acc_one_hot = eval(salt_and_pepper_one_hot[1])

fgsm_one_hot = fgsm_f.read().split("\n")
fgsm_epsilons_one_hot = eval(fgsm_one_hot[0])
fgsm_acc_one_hot = eval(fgsm_one_hot[1])

fgsm_f.close()
salt_and_pepper_f.close()
gaussian_blur_f.close()
gaussian_noise_f.close()
negative_f.close()
```

```python
## ECOC multi-hot encoding 결과 불러오기
```

```python
negative_f = open("result/Negative(ECOC).txt", "r")
gaussian_noise_f = open("result/GaussianNoise(ECOC).txt", "r")
gaussian_blur_f = open("result/GaussianBlur(ECOC).txt", "r")
salt_and_pepper_f = open("result/SaltAndPepper(ECOC).txt", "r")
fgsm_f = open("result/FGSM(ECOC).txt", "r")

negative_ecoc = negative_f.read().split("\n")
negative_acc_ecoc = eval(negative_ecoc[0])

gaussian_noise_ecoc = gaussian_noise_f.read().split("\n")
gaussian_noise_var_ecoc = eval(gaussian_noise_ecoc[0])
gaussian_noise_acc_ecoc = eval(gaussian_noise_ecoc[1])

gaussian_blur_ecoc = gaussian_blur_f.read().split("\n")
gaussian_blur_sigma_ecoc = eval(gaussian_blur_ecoc[0])
gaussian_blur_acc_ecoc = eval(gaussian_blur_ecoc[1])

salt_and_pepper_ecoc = salt_and_pepper_f.read().split("\n")
salt_and_pepper_amount_ecoc = eval(salt_and_pepper_ecoc[0])
salt_and_pepper_acc_ecoc = eval(salt_and_pepper_ecoc[1])

fgsm_ecoc = fgsm_f.read().split("\n")
fgsm_epsilons_ecoc = eval(fgsm_ecoc[0])
fgsm_acc_ecoc = eval(fgsm_ecoc[1])

fgsm_f.close()
salt_and_pepper_f.close()
gaussian_blur_f.close()
gaussian_noise_f.close()
negative_f.close()
```

```python
## negative
```

```python
fig, ax1 = plt.subplots()

plt.title('Negative noise')
ax1.set_ylabel('Accuracy')
ax1.axis([-1, 3, 0, 100])
ax1.grid()
ax1.get_xaxis().set_visible(False)

plt.plot([0, 1, 2], [negative_acc_one_hot, negative_acc_one_hot, negative_acc_one_hot], "gp", label = "One-Hot")
plt.plot([0, 1, 2], [negative_acc_one_hot, negative_acc_one_hot, negative_acc_one_hot], color = "g")
plt.plot([0, 1, 2], [negative_acc_ecoc, negative_acc_ecoc, negative_acc_ecoc], "r*", label = "ECOC")
plt.plot([0, 1, 2], [negative_acc_ecoc, negative_acc_ecoc, negative_acc_ecoc], color = "r")

plt.legend(loc='upper right')
```

```python
## Gaussian noise
```

```python
fig, ax1 = plt.subplots()

plt.title('Gaussian noise')
ax1.set_xlabel('var')
ax1.set_ylabel('Accuracy')
ax1.axis([0, gaussian_noise_var_one_hot[-1], 0, 100])
ax1.grid()

plt.plot(gaussian_noise_var_one_hot, gaussian_noise_acc_one_hot, "gp", label = "One-Hot")
plt.plot(gaussian_noise_var_one_hot, gaussian_noise_acc_one_hot, color = "g")
plt.plot(gaussian_noise_var_ecoc, gaussian_noise_acc_ecoc, "r*", label = "ECOC")
plt.plot(gaussian_noise_var_ecoc, gaussian_noise_acc_ecoc, color = "r")

plt.legend(loc='upper right')
```

```python
## Gaussian blur
```

```python
fig, ax1 = plt.subplots()

plt.title('Gaussian blur')
ax1.set_xlabel('sigma')
ax1.set_ylabel('Accuracy')
ax1.axis([gaussian_blur_sigma_one_hot[-1], gaussian_blur_sigma_one_hot[0], 0, 100])
ax1.grid()

plt.plot(gaussian_blur_sigma_one_hot, gaussian_blur_acc_one_hot, "gp", label = "One-Hot")
plt.plot(gaussian_blur_sigma_one_hot, gaussian_blur_acc_one_hot, color = "g")
plt.plot(gaussian_blur_sigma_ecoc, gaussian_blur_acc_ecoc, "r*", label = "ECOC")
plt.plot(gaussian_blur_sigma_ecoc, gaussian_blur_acc_ecoc, color = "r")

plt.legend(loc='upper right')
```

```python
fig, ax1 = plt.subplots()

plt.title('Salt and Pepper')
ax1.set_xlabel('amount')
ax1.set_ylabel('Accuracy')
ax1.axis([0, salt_and_pepper_amount_one_hot[-1], 0, 100])
ax1.grid()

plt.plot(salt_and_pepper_amount_one_hot, salt_and_pepper_acc_one_hot, "gp", label = "One-Hot")
plt.plot(salt_and_pepper_amount_one_hot, salt_and_pepper_acc_one_hot, color = "g")
plt.plot(salt_and_pepper_amount_ecoc, salt_and_pepper_acc_ecoc, "r*", label = "ECOC")
plt.plot(salt_and_pepper_amount_ecoc, salt_and_pepper_acc_ecoc, color = "r")

plt.legend(loc='upper right')
```
